```python
import requests
import re
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
links = [
    'https://sec.report/Document/0001193125-21-072571/',
    'https://sec.report/Document/0001193125-21-031547/',
    'https://sec.report/Document/0001193125-21-024746/',
    'https://sec.report/Document/0001193125-21-056708/'
]
```

# Link 1
Data to parse:<br>
IPO Price: 11.50\\$ <br>
Number Shares: 12,218,750 <br>
IPO Unit: 10\\$

```python
driver = webdriver.Chrome(r'D:\Programming\Projects\TripAdvisorRestaurantRating\data\chromedriver.exe')
driver.get(links[0])

html = driver.page_source
```

```python
soup = BeautifulSoup(html)
```

```python
doc_html = soup.find('div', {'class': 'html-body'})
for m in re.finditer(r"(\s\$(11\.50)\s)|(\s\$(11\.50))", doc_html.text):
    print('%02d-%02d: %s' % (m.start(), m.end(), m.group(0)))
```

```python
for m in re.finditer(r"(\s(12,218,750)\s)", doc_html.text):
    print('%02d-%02d: %s' % (m.start(), m.end(), m.group(0)))
```

```python
for m in re.finditer(r"(.\$(10\.00).)", doc_html.text):
    print('%02d-%02d: %s' % (m.start(), m.end(), m.group(0)))
```

```python
driver.close()
driver.quit()
```
