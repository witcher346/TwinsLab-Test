```python
import requests
import regex as re
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
links = [
    'https://sec.report/Document/0001193125-21-072571/',
    'https://sec.report/Document/0001193125-21-031547/',
    'https://sec.report/Document/0001193125-21-024746/',
    'https://sec.report/Document/0001193125-21-056708/'
]
```

# Link 1
Data to parse:<br>
IPO Price: 11.50\\$ <br>
Number Shares: 12,218,750 <br>
IPO Unit: 10\\$

```python
driver = webdriver.Chrome(r'D:\Programming\Projects\TripAdvisorRestaurantRating\data\chromedriver.exe')
driver.get(links[0])

html = driver.page_source
```

```python
soup = BeautifulSoup(html)
```

```python
doc_html = soup.find('div', {'class': 'html-body'})
for m in re.finditer(r"(\$\d\d\.\d\d\sper share)", doc_html.text):
    print('%02d-%02d: %s' % (m.start(), m.end(), m.group(0)))
```

```python
for m in re.finditer(r"(\s\d\d,\d+,\d+(?<=units.+))", doc_html.text):
    print('%02d-%02d: %s' % (m.start(), m.end(), m.group(0)))
```

```python
search = re.search(r'initial offering price.+', doc_html.text)

sentence_end_index = re.search(r'\.\s', search.group()).end()
sentence = search.group()[:sentence_end_index]
sentence
```

```python
re.search(r'(?<=\sof\s).*(?=per)', sentence).group()
```

```python
driver.close()
driver.quit()
```
